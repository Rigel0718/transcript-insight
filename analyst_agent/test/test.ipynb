```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import sys
import os 

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../../")))
```

```python
from base_node import RunLogger
from base_node import Env

logger = RunLogger(base_name='transcript_agent')

from datetime import datetime
env = Env(
    work_dir='./result/',
    user_id='TEST_02',
    run_logger=logger
)

run_id = datetime.now().strftime("%Y-%m-%d-%H-%M%S")

from langchain_core.runnables import RunnableConfig  
config = RunnableConfig(thread_id=run_id, max_iterations=40)

print("config Runtime ENV")
```

```python
from analyst_agent import transcript_analyst_graph, ReportState, AnalysisSpec

graph = transcript_analyst_graph(verbose=True, env=env, track_time=True)
print('init transcript_analyst_graph')
```

```python
# Load DATA 

import json
file_path = './data/virtual_data01.json'
with open(file_path, 'r', encoding='utf-8') as f:
    data_dict = json.load(f)

text_for_llm = json.dumps(data_dict, ensure_ascii=False, indent=2)

print('Success load data')
```

```python
ai_recruiter = AnalysisSpec(
    focus=["GPA trend","major GPA","math-related courses"],
    audience="evaluator",
    audience_spec="AI company recruiter",
    audience_goal="수학적 사고가 뛰어난 지원자를 선발",
    audience_values=["성실성","논리적 문제 해결","성장 가능성"],
    evaluation_criteria=["전공 과목 성취도","수학 과목 성적","꾸준한 GPA 상승"],
    decision_context="채용 선발",
    
    time_scope="전체 학기",
    comparison_target="동일 전공 평균",
    priority_focus=["전공 GPA", "수학적 과목 성취도"],
    
    tone="formal",
    language="ko",
    detail_level="in_depth",
    insight_style="comparative",
    evidence_emphasis="high",
    tone_variation="간결하지만 설득력 있게",
    
    output_format=["text","chart","recommendation"],
    include_recommendations=True,
    highlight_style="strengths"
)

print("Success init AnalysisSpec -Example AI Recruiter-")
```

```python
input = ReportState(
    # user_input
    user_query='',
    dataset=text_for_llm,
    run_id=run_id,

    # report_element
    report_plan=[],
    metric_plan=[],
    inform_metric='',   
    analyst=ai_recruiter,         

    rewrite_query='',

    # etc
    cost=0.0,
)

print("Success init ReportState")
```

```python
final_result = graph.invoke(input = input, config = config)
```

```python
final_report = final_result['report']
```

```python
file_path = f'./result/users/TEST_02/{run_id}.md'
with open(file_path, "w", encoding="utf-8") as f:
    f.write(final_report)
```
